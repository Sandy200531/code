# Advanced Pandas Concepts

### Grouping and Aggrigation in pandas

Grouping and aggregation are powerful operations in Pandas for summarizing and analyzing data. Here are some illustrative examples of grouping and aggregation in Pandas

some techniques for grouping -   
1. Grouping single column and apply inbuilt aggrigation function.

```python
import pandas as pd

data = {
    'Product': ['A', 'B', 'A', 'C', 'B'],
    'Sales': [100, 200, 150, 120, 250]
}

df = pd.DataFrame(data)



```

```python
df
```

```python
df.groupby('Product')
```

```python
for i,j in df.groupby('Product'):
    print(f'product name:{i}')
    print(f'DataFrame:\n{j}')
```

```python
df.groupby('Product')['Sales'].min()
```

```python
df.groupby('Product')['Sales'].size()
```

```python
# Group by 'Product' and calculate the total sales for each category
result = df.groupby('Product')['Sales'].sum()
print(result)
```

2. Grouping by Multiple Columns and Aggregating

```python
import pandas as pd

data = {
    'Department': ['HR', 'IT', 'HR', 'IT', 'Finance'],
    'JobTitle': ['Manager', 'Developer', 'Coordinator', 'Developer', 'Analyst'],
    'Salary': [80000, 90000, 60000, 95000, 75000]
}

df = pd.DataFrame(data)


```

```python
df
```

```python
from IPython.display import display
```

```python
for i,j in df.groupby(['Department', 'JobTitle']):
    print(f'product name:{i}')
    display(j)
    print('\n\n')
```

```python
df.groupby(['Department', 'JobTitle'])['Salary'].mean()
```

```python
# Group by 'Department' and 'JobTitle' and calculate the average salary
result = df.groupby(['Department', 'JobTitle'])['Salary'].mean()
print(result)
```

### 3. Apply custom function

```python
data = {
    'Product': ['A', 'B', 'A', 'C', 'B'],
    'Sales': [100, 200, 150, 120, 250]
}

df = pd.DataFrame(data)

```

```python
def data_range(series):
    return series.max() + series.min()
```

```python
for i,j in df.groupby('Product'):
    print(f'product name:{i}')
    display(j)
    print('\n\n')
```

```python
df.groupby('Product')['Sales'].apply(data_range)
```

```python
# Define a custom aggregation function to calculate the range
def data_range(series):
    return series.max() - series.min()

# Group by 'Product' and apply the custom aggregation function
result = df.groupby('Product')['Sales'].apply(data_range)
print(result)
```

### Data Visualization in pandas

### Line Plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create a DataFrame with some sample data
data = {'Time': np.arange(0, 10, 0.1), 'Value': np.sin(np.arange(0, 10, 0.1))}
df = pd.DataFrame(data)
```

```python
df.columns=['col1','col2']
```

```python
df
```

```python


# Create a line plot
df.plot(x='col1', y='col2', title='Line Plot')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()

```

### Bar Plot

```python
# Create a bar plot
data = {'Category': ['A', 'B', 'C', 'D'], 'Value': [10, 25, 15, 30]}
df = pd.DataFrame(data)

df.plot(x='Category', y='Value', kind='bar', title='Bar Plot')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

### Scatter Plot

```python
# Create a scatter plot
data = {'X': np.random.rand(50), 'Y': np.random.rand(50)}
df = pd.DataFrame(data)

df.plot(x='X', y='Y', kind='scatter', title='Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

### Histogram

```python
# Create a histogram
data = {'Values': np.random.randn(1000)}
df = pd.DataFrame(data)

df['Values'].plot(kind='hist', bins=100, title='Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
